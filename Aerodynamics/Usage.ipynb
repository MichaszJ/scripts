# Aerodynamics Scripts - Usage
## Flight Envelope Plotter

Generates a plot of an aircraft's flight envelope based on its flight and structural properties

```python
import matplotlib.pyplot as plt
from aerodynamics import flight_envelope_graph
```

```python
# given properties
cruise_mach = 0.80
cruise_alt = 12000

max_mach = 0.85
max_cl = 1.5

max_takeoff = 48000
wing_area = 90.0

n_lim = 2.1 + 24000 / (max_takeoff * 2.20462 + 10000)
```

```python
plt.figure(dpi=230)

flight_envelope_graph(n_lim, wing_area, max_takeoff, cruise_mach, cruise_alt, max_mach, max_cl)

plt.grid(linestyle='--')
plt.ylim([-1.5, 3])
plt.xlim([-5, 155])

plt.show()
```
